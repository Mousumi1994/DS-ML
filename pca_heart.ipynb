```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#loading dataset
df = pd.read_csv('heart.csv')
df.head()
```

```python
#counting null
df.isna().sum()
```

```python
#outlier by z-score
```

```python
age = df['Age']
age_mean = np.mean(age)
age_std = np.std(age)
threshold = 3
outlier_age = []
for i in age:
    z = (i-age_mean)/age_std
    if z > threshold:
        outlier_age.append(i)
outlier_age
```

```python
restingBP = df['RestingBP']
restingBP_mean = np.mean(restingBP)
restingBP_std = np.std(restingBP)
threshold = 3
outlier_restingBP = []
for i in restingBP:
    z = (i-restingBP_mean)/restingBP_std
    if z > threshold:
        outlier_restingBP.append(i)
outlier_restingBP
```

```python
cholesterol = df['Cholesterol']
cholesterol_mean = np.mean(cholesterol)
cholesterol_std = np.std(cholesterol)
threshold = 3
outlier_cholesterol = []
for i in cholesterol:
    z = (i-cholesterol_mean)/cholesterol_std
    if z > threshold:
        outlier_cholesterol.append(i)
outlier_cholesterol
```

```python
fastingBS = df['FastingBS']
fastingBS_mean = np.mean(fastingBS)
fastingBS_std = np.std(fastingBS)
threshold = 3
outlier_fastingBS = []
for i in fastingBS:
    z = (i-fastingBS_mean)/fastingBS_std
    if z > threshold:
        outlier_fastingBS.append(i)
outlier_fastingBS
```

```python
maxHR = df['MaxHR']
maxHR_mean = np.mean(maxHR)
maxHR_std = np.std(maxHR)
threshold = 3
outlier_maxHR = []
for i in maxHR:
    z = (i-maxHR_mean)/maxHR_std
    if z > threshold:
        outlier_maxHR.append(i)
outlier_maxHR
```

```python
oldpeak = df['Oldpeak']
oldpeak_mean = np.mean(oldpeak)
oldpeak_std = np.std(oldpeak)
threshold = 3
outlier_oldpeak = []
for i in oldpeak:
    z = (i-oldpeak_mean)/oldpeak_std
    if z > threshold:
        outlier_oldpeak.append(i)
outlier_oldpeak
```

```python
heartDisease = df['HeartDisease']
heartDisease_mean = np.mean(heartDisease)
heartDisease_std = np.std(heartDisease)
threshold = 3
outlier_heartDisease = []
for i in heartDisease:
    z = (i-heartDisease_mean)/heartDisease_std
    if z > threshold:
        outlier_heartDisease.append(i)
outlier_heartDisease
```

```python
df.shape
```

```python
#removing outlier
```

```python
i = df[df['RestingBP'].isin(outlier_restingBP)].index
df.drop(i, inplace=True)
df.shape
```

```python
i = df[df['Cholesterol'].isin(outlier_cholesterol)].index
df.drop(i, inplace=True)
df.shape
```

```python
i = df[df['Oldpeak'].isin(outlier_oldpeak)].index
df.drop(i, inplace=True)
df.shape
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df.describe()
```

```python
x = df.drop('HeartDisease', axis=1)
x.head()
```

```python
y = df['HeartDisease']
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.2, random_state=2)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
l_reg = LogisticRegression(C=0.5)
s_clf = SVC(C=0.5, kernel='rbf', gamma='auto')
r_clf = RandomForestClassifier(n_estimators=100, max_depth=10, max_samples=50, random_state=20)
```

```python
l_reg.fit(x_train, y_train)
l_reg.score(x_test, y_test)
```

```python
s_clf.fit(x_train, y_train)
s_clf.score(x_test, y_test)
```

```python
r_clf.fit(x_train, y_train)
r_clf.score(x_test, y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(x)
x_pca = pca.transform(x)
x_pca
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=2)
```

```python
l_reg.fit(x_train, y_train)
l_reg.score(x_test, y_test)
```

```python
s_clf.fit(x_train, y_train)
s_clf.score(x_test, y_test)
```

```python
r_clf.fit(x_train, y_train)
r_clf.score(x_test, y_test)
```
