```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

###### Importing CSV file

```python
Customer = pd.read_csv("Mall_Customers.csv")
Customer.head()
```

```python
## print shape of dataset with rows and columns
Customer.shape
```

```python
#Get brief of data
Customer.info()
```

```python
sns.barplot(x = 'Gender', y = 'Spending Score (1-100)', data = Customer)
```

Female customers' Spending Score is higher than Male Customer

```python
sns.set(font_scale=1)
sns.scatterplot(x = 'Age', y = 'Spending Score (1-100)', data = Customer)
```

Customers whose age is under 40 have higher Spending Score

```python
sns.countplot(x ='Gender', data = Customer)
```

```python
sns.set(font_scale=1)
sns.histplot(data = Customer['Spending Score (1-100)'], kde = True)
```
